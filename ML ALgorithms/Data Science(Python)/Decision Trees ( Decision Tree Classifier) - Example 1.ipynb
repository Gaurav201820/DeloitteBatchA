```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_decisiontree = pd.read_csv('C:/Deloitte Folder/ML/Iris.csv')
```

```python
df_decisiontree.head()
```

```python
x=df_decisiontree.iloc[:,3:5].values
y=df_decisiontree.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeClassifier
# by default the criterion is 'gini index'. we can change to entropy
trees=DecisionTreeClassifier(random_state=12) 
trees.fit(x,y)
```
