bar plot and count plot is plotted on categorical data

CATEGORICAL DATA PLOTTING

    1. BAR PLOT
    
    2. COUNTPLOT    
 
    3. SWARMPLOT
    
    4. STRIPPLOT(ONE CATEGORICAL VALUE AND ONE NUMERICAL VALUE)
    
    5. BOX PLOT
    
    6. VIOLIN PLOT

```python
#barplot
sns.barplot(data=df,y=df['total_bill'],x=df['sex'],estimator = np.std) #estimator takes a vlue and plots the graph according 
                                                                     #to the value of estimator. the ranges of y axis changes 
                                                                #if estimator is specified. otherwise deafultly it will be chosen
```

```python
sns.violinplot(data=df,y=df['total_bill'],x=df['sex'],estimator = np.std)
```

```python
#countplot:- plots the graph based on the count of values in the column. here in this case count of the gender 
sns.countplot(x=df['sex'],data=df)
```

```python
sns.countplot(x=df['size'],data=df)
```

```python
sns.boxplot(data=df,x=df['total_bill'])
```

```python
sns.boxplot(data=df,y=df['tip'],x=df['total_bill']) 
# boxplot isnt useful this type of data. its helpful in numerical(categorical)datalike only one column                                                  data
                                                    
```

```python
sns.boxplot(data=df,palette='rainbow')#box plot takes only numerical values.
#palette specifies the color
```

```python
sns.boxplot(data=df,x='day',y='total_bill',hue='sex',palette='coolwarm')
# plotting boxplot on 3 variables (x= day,y=total bill, data(hue)=sex)
```

    SWARM PLOT

```python
sns.swarmplot(x='total_bill',y='day',data = df)
```

        STRIP PLOT

```python
sns.stripplot(x='day',y='total_bill',data=df) 
#stripplot uses one categorical and one numerical value
#it plots based on x and y axis. it doesnt differentiate in the data. to do so use hue parameter
```

```python
sns.stripplot(data=df,x='day',y='total_bill',hue='sex',palette='Set1')
#hue is to differntiate in the data(plotting area as shown in the below graph) with respect to 2 other variables(x and y axis)
```

    DISTRIBUTION PLOTS
    
    1. DISTPLOT(DEPRICATED VERSION)
    
    2. DISPLOT(SAME AS DISTPLOT(NEW VERSION))
    
    3.JOINTPLOT
    
    4. PAIRPLOT

```python
sns.distplot(df['total_bill'],kde=False)
#kde =false emilinates the bell shaped curve. by default its true for distplot(depricated version)
```

```python
sns.displot(df['total_bill'],kde=True)
#kde =true brings up the bell shaped curve. by default its false for displot
```

```python
sns.jointplot(data=df,x='total_bill',y='tip',hue='sex')
```

Pair plot plots the numerical value columns in a dataset pair wise. it plots a column with each other columns

```python
sns.pairplot(df)
#for univariate data that is totalbill and total bill for x and y axis, we get distribution plots.
```

```python
sns.pairplot(data=df,hue='sex',palette = 'coolwarm')
```

```python
sns.pairplot(data=df,hue='smoker')
```
