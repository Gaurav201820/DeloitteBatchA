```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logregressor=LogisticRegression()
logregressor.fit(x_train,y_train)
```

```python
y_pred = logregressor.predict(x_test)
```

```python
y_pred
```

### Using Confusion Matrix to measure the accuracy of the logistic regression model

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

#### classification report used to measure the precision and recall

Precision tells what proportion of the acutal values was correct

Recall tells what proportion of the actual values was predicted correctly by our model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
