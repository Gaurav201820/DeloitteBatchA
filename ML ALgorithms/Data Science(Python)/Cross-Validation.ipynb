```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('C:/Deloitte Folder/ML/Logistic_advertising.csv')
```

```python
data.head()
```

```python
data.drop(['Ad Topic Line','City','Country','Timestamp'],axis=1,inplace=True)
```

```python
data.head()
```

### Manually Cross validating

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
k=5
kf = KFold(n_splits=k,random_state=None)
model = LogisticRegression()
acc_score=[]
```

```python
list(kf.split(x))
```

```python
for train_index,test_index in kf.split(x):
    x_train,x_test = x.iloc[train_index,:],x.iloc[test_index,:]
    y_train,y_test = y[train_index],y[test_index]
    
    model.fit(x_train,y_train)
    predictions = model.predict(x_test)
    acc=accuracy_score(predictions,y_test)
    acc_score.append(acc)
```

```python
print(acc_score)
```

```python
avg_acc_score=sum(acc_score)/k
avg_acc_score
```

### Cross validation using cross_val_score method from python

```python
from sklearn.model_selection import cross_val_score
```

```python
result = cross_val_score(model,x,y,cv=kf)
print(result.mean()*100)
```

### Cross validation using LogisticregressionCV(built in CV method)

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score
clt=LogisticRegressionCV()
clt.fit(x_train,y_train)
t_pred = clt.predict(x_test)
acc = accuracy_score(y_test,t_pred)
print(acc*100)
```
