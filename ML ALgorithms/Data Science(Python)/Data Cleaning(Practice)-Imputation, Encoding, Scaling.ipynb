```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('C:/Deloitte Folder/Admission_Predict.csv')
df.head()
```

```python
df.pop(0)
```

```python
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state=1)
#random state controls the shuffling of dataset for test and train everytime the set is run.
#if random state = value then everytime the the program is run whoever runs the program gets the same splits everytime
print('x_train:',x_train)
print('x_test:',x_test)
print('y_train:',y_train)
print('y_test:',y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print('x train:',x_train)
print('x test:',x_test)
```
